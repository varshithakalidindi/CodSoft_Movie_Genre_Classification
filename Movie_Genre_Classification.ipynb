```python
import pandas as pd

# Load training data
train_data = pd.read_csv(
    'train_data.txt',
    sep=':::',
    engine='python',
    names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION']
)

# Display first 5 rows
train_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Create TF-IDF + Logistic Regression pipeline
model = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=5000)),
    ('classifier', LogisticRegression(max_iter=1000))
])

# Train the model
model.fit(train_data['DESCRIPTION'], train_data['GENRE'])

print("Model trained successfully!")
```

```python
# Test with a sample movie description
sample_description = ["A young wizard discovers his magical powers and fights dark forces to save the world."]

prediction = model.predict(sample_description)

print("Predicted Genre:", prediction[0])
```

```python
# Load test data
test_data = pd.read_csv(
    'test_data_solution.txt',
    sep=':::',
    engine='python',
    names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION']
)

# Predict genres for test data
predictions = model.predict(test_data['DESCRIPTION'])

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test_data['GENRE'], predictions)

print("Accuracy:", accuracy)
```

```python
# User input prediction
movie_description = input("Enter movie description: ")

prediction = model.predict([movie_description])

print("Predicted Genre:", prediction[0])
```
